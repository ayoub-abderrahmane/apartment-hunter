```python
import pandas as pd

# Import de pandas et chargement du df avec affichage des 5 premières lignes

df = pd.read_csv("bdd/houses_Madrid.csv")
df.head()
```

```python
# Affichage des valeurs manquantes
df.isnull().sum()
```

```python
# Montre la taille du dataset
print(df.shape)
```

```python
# Supression des colonnes vides et quasiment vides

df = df.drop(["latitude", "longitude", "portal", "door", "rent_price_by_area", "n_floors", "sq_mt_allotment", "sq_mt_useful", "street_number", "built_year" ,"is_orientation_north" ,"is_orientation_west","is_orientation_south","is_orientation_east","has_public_parking","has_private_parking" ,"are_pets_allowed" ,'is_parking_included_in_price' ,"has_ac" ,"has_green_zones" ,"is_accessible" ,"is_kitchen_equipped","is_furnished","has_storage_room","has_balcony","has_terrace","has_pool","has_garden","raw_address","street_name","has_central_heating","has_individual_heating","has_fitted_wardrobes", "parking_price"], axis=1) 
```

```python


df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

cols_bool = ['is_floor_under', 'is_new_development', 'has_lift', 'is_exterior', 'house_type_id']  # Colonne de booléens

imputer_bool = SimpleImputer(strategy='most_frequent')
df[cols_bool] = imputer_bool.fit_transform(df[cols_bool])
```
